```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the TSV file into a DataFrame
file_path = 'D://Chrome Downloads/Climate Analysis Final - climate_analysis.tsv'
df = pd.read_csv(file_path, sep='\t')

# Count the number of rows for each phylum
phylum_counts = df['Phylum'].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))
phylum_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Samples by Phylum')
plt.xlabel('Phylum')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
